```python
# Importing Libraries
import os
from PIL import Image
import numpy as np


# Setting Paths
dir_data = '/content/drive/MyDrive/Colab/Detecting_and_Classifying_Car_Damage/datasets_gray_npy'
category_damages = ['scratch']
category_data = ['train']
category_represent = ['images']
category_represents = ['masks']

ind = []

# Making Directories and Changing to npz files
for damage in category_damages:
    for data in category_data:
        for represent in category_represent:
            cnt = 0
            for file in os.listdir(os.path.join(dir_data, damage, data, represent)):
                ind.append(file)
                # d = np.load(os.path.join(dir_data, damage, data, represent, file))
                # if d.shape != (512,512):
                #   print(file)
                cnt += 1
            print('{0}_{1}_{2}: {3} done'.format(damage, data, represent, cnt))

# Making Directories and Changing to npz files
for damage in category_damages:
    for data in category_data:
        for represent in category_represents:
            cnt = 0
            for file in os.listdir(os.path.join(dir_data, damage, data, represent)):
                if file in ind:
                    ind.remove(file)
                # d = np.load(os.path.join(dir_data, damage, data, represent, file))
                # if d.shape != (512,512):
                #   print(file)
                cnt += 1
            print('{0}_{1}_{2}: {3} done'.format(damage, data, represent, cnt))

print(ind)

print('All Done!')
```
